```python
import pandas as pd
```

```python
# 读取 kg.csv
df = pd.read_csv('kg.csv')
```

```python

# 打印列名
print("列名：")
print(df.columns.tolist())

# 打印前几行
print("\n前5行数据：")
print(df.head())
```

```python
print("relation 列的唯一值：")
print(df['relation'].unique())
print(f"\n总共有 {df['relation'].nunique()} 种不同的关系")
print("\nx_type 的唯一值：")
print(df['x_type'].unique())

print("\ny_type 的唯一值：")
print(df['y_type'].unique())
```

```python
# 筛选 disease → protein 的边
disease_protein_edges = df[
    (df['relation'] == 'disease_protein') &
    (df['x_type'] == 'disease') &
    (df['y_type'] == 'gene/protein')
].copy()

print(f"共提取到 {len(disease_protein_edges)} 条 Disease → Protein 边")

# 打印前10条看看
print("\n前10条 Disease-Protein 边：")
disease_protein_edges
```

```python
disease_protein_edges = disease_protein_edges.drop_duplicates(
    subset=['x_id', 'y_id']
)
disease_protein_edges
```
